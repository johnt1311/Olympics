```python
# load medal counts
medal_counts_country = pd.read_csv("Completed Resources\complete_country_medals (7-29-2021).csv")
medal_counts_country.head()
```

```python
# import world bank GDP
world_bank_gdp = pd.read_csv(fr"{folder}\GDP_Data_WorldBank.csv")
world_bank_gdp.head()
```

```python
# drop columns we don't need
world_bank_gdp.drop(columns=["Unnamed: 56", "From Medal Count", "On This Sheet?"], inplace=True)
# world_bank_gdp.drop(columns=["Unnamed: 55"], inplace=True)
world_bank_gdp.head()
```

```python
# drop the NaN rows
world_bank_gdp.dropna(axis=0, inplace=True)
world_bank_gdp.head()
```

```python
# replace all the '..' with 0
world_bank_gdp.replace('\.{2}', 0, regex=True, inplace=True)
world_bank_gdp.head()
```

```python
# dtypes are objects so convert to numeric
GDP_columns = world_bank_gdp.columns.to_list()
# dump the first 2 columns they are ok as objects
GDP_columns = GDP_columns[2:]
# use this list for df.apply()
world_bank_gdp[GDP_columns] = world_bank_gdp[GDP_columns].apply(pd.to_numeric)
world_bank_gdp.info()

# cell output hidden with 'o'
```

```python
world_bank_gdp.to_csv(fr"{folder}\07-30-2021_World_Bank_GDP.csv", index=False)
```
