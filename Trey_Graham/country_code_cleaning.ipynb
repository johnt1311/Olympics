```python
import numpy as np
```

```python
# create function to add the rest of the countries
# get list of the years we want
gdp_years = gdp_layout["Year"].to_list()

gdp_list = []
country_code_list = []
df_years = []
# pull every row and use the column index (will add steps later for function)
for i in range(0, len(GDP_test.columns.to_list()), 2):
    # create list of all of the GDP data
    gdp_list.append(GDP_test.iloc[:, i].values)

for i in range(1, len(GDP_test.columns.to_list()), 2):
    # create list of all of the GDP data
    country_code_list.append(GDP_test.iloc[:, i].values)

for i in range(0, len(GDP_test.columns.to_list()), 2):
    # create list of all of the GDP data
    df_years.append(gdp_years)

# create dataframe from the lists #NOTE these were not working with the data as a list of arrays - did not try with the unpacked lists. 
# formatted_gdp_data = pd.DataFrame(np.column_stack([df_years, country_code_list, gdp_list]), columns=['Year', 'Country Code', 'GDP'])
# formatted_gdp_data = pd.DataFrame(np.concatenate(df_years, country_code_list, gdp_list))

# flatten the lists - double list comprehension yo
def flatten(x):
    return [item for sublist in x for item in sublist]

gdp_list = flatten(gdp_list)
country_code_list = flatten(country_code_list)
df_years = flatten(df_years)

# create the dataframe by zipping the lists
formatted_gdp_data = pd.DataFrame(zip(df_years, country_code_list, gdp_list),\
    columns=['Year', 'Country Code', 'GDP'])


print(formatted_gdp_data.shape)
formatted_gdp_data.head()
```
